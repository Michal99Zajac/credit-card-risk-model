```python
# import packages/modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load CSV file into a pandas dataframe
df = pd.read_csv('../../data/orginal/credit_card_approval.csv')
```

# CODE_GENDER

```python
# Create a plot from the "GENDER_CODE" column
counts = df['CODE_GENDER'].value_counts(dropna=False)
counts.plot(kind='pie', labels=counts.index, autopct='%1.1f%%')

# Add a title to the plot
plt.title('Gender Distribution')

# Show the plot
plt.savefig('../../media/data_visual/CODE_GENDER__PIE.png')
```

# AMT_INCOME_TOTAL

```python
# Create a plot from the "AMT_INCOME_TOTAL" column
totals = df['AMT_INCOME_TOTAL']
totals.plot(kind='hist', bins=30)

# Add a title and x-axis label to the plot
plt.title('Distribution of Annual Income')
plt.xlabel('Annual Income')

# Show the plot
plt.savefig('../../media/data_visual/AMT_INCOME_TOTAL__HIST.png')
```

```python
fig, ax = plt.subplots()
df.boxplot(ax=ax, column="AMT_INCOME_TOTAL")

ax.set_yscale('log')

ax.set_ylabel('Value (log scale)')
ax.set_title('Annual Income Boxplot')

plt.savefig('../../media/data_visual/AMT_INCOME_TOTAL__BOXPLOT.png')
```

# NAME_EDUCATION_TYPE

```python
# Create a plot from the "NAME_EDUCATION_TYPE" column
counts = df['NAME_EDUCATION_TYPE'].value_counts(dropna=False)
ax = counts.plot(kind='pie', labels=None)

# Add a title to the plot
plt.title('Education Level Distribution')

# Calculate percentage values
percentages = (counts / counts.sum() * 100).round(2).astype(str) + '%'

# Create a list of labels with percentage values
labels_with_percentages = [f'{label} ({percentage})' for label, percentage in zip(counts.index, percentages)]

# Set the legend title and position
ax.legend(title='Education Levels', labels=labels_with_percentages, bbox_to_anchor=(0, -0.4), loc='lower left', borderaxespad=0)

# Show the plot
plt.savefig('../../media/data_visual/NAME_EDUCATION_TYPE__PIE.png')
```
